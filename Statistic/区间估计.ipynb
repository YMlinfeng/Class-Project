# 方差已知

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
np.random.seed()
n=100
samples=[np.random.normal(loc=0,scale=3,size=10)for _ in range(n)]
fig,ax=plt.subplots(figsize=(10,7))
for i in np.arange(1,n,1):
    sample_mean=np.mean(samples[i])
    se=3/np.sqrt(len(samples[i]))
    h=se*stats.norm.ppf((1+0.95)/2)
    sample_ci=[sample_mean - h,sample_mean + h]
    
    if ((sample_ci[0]<=0)and(0<=sample_ci[1])):
        #print(sample_ci[0],sample_ci[1])
        plt.plot((i,i),(sample_ci[0],sample_ci[1]),color='blue',linewidth=1)
        #plt.plot(np.mean(samples[i]),i,'bo')
    else:
        plt.plot((i,i),(sample_ci[0],sample_ci[1]),color='red',linewidth=1)
        #plt.plot(np.mean(samples[i]),i,'ro')
plt.axhline(y=0,xmin=0,xmax=1,linestyle='--')
#plt.legend(loc='best')
```

# 方差未知

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
np.random.seed()
n=100
samples=[np.random.normal(loc=0,scale=3,size=30)for _ in range(n)]
fig,ax=plt.subplots(figsize=(10,7))
for i in np.arange(1,n,1):
    sample_mean=np.mean(samples[i])
    #计算标准差
    se=stats.sem(samples[i])
    h=se*stats.t.ppf((1+0.95)/2,len(samples[i])-1)
    sample_ci=[sample_mean - h,sample_mean + h]
    
    if ((sample_ci[0]<=0)and(0<=sample_ci[1])):
        #print(sample_ci[0],sample_ci[1])
        plt.plot((i,i),(sample_ci[0],sample_ci[1]),color='blue',linewidth=1)
        #plt.plot(np.mean(samples[i]),i,'bo')
    else:
        plt.plot((i,i),(sample_ci[0],sample_ci[1]),color='red',linewidth=1)
        #plt.plot(np.mean(samples[i]),i,'ro')
plt.axhline(y=0,xmin=0,xmax=1,linestyle='--')
#plt.legend(loc='best')
```
